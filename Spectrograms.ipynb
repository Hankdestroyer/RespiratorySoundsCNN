```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Directorio principal que contiene las carpetas para cada tipo
directorio_principal = './Input'

# Directorio donde se guardarán los espectrogramas
directorio_espectrogramas = './Spectrograms'

# Crear las carpetas de espectrogramas si no existen
tipos = {
    '00': 'normal',
    '10': 'crackle',
    '01': 'wheeze',
    '11': 'crackle_wheeze'
}
for tipo in tipos.values():
    carpeta_tipo = os.path.join(directorio_espectrogramas, tipo)
    os.makedirs(carpeta_tipo, exist_ok=True)

# Iterar sobre las carpetas para cada tipo
for tipo_codigo, tipo_texto in tipos.items():
    carpeta_tipo = os.path.join(directorio_principal, tipo_texto)
    
    # Iterar sobre los archivos .wav en la carpeta del tipo
    for filename in os.listdir(carpeta_tipo):
        if filename.endswith('.wav'):
            # Ruta del archivo de audio
            audio_path = os.path.join(carpeta_tipo, filename)
            
            # Cargar el archivo de audio
            y, sr = librosa.load(audio_path)
            
            # Calcular el espectrograma mel
            mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
            
            # Convertir a escala logarítmica (dB)
            mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)
            
            # Ruta para guardar el espectrograma
            carpeta_espectrograma = os.path.join(directorio_espectrogramas, tipo_texto)
            save_file = os.path.join(carpeta_espectrograma, os.path.splitext(filename)[0] + '.png')
            
            # Guardar el espectrograma
            plt.figure(figsize=(10, 4))
            librosa.display.specshow(mel_spectrogram_db, sr=sr, x_axis='time', y_axis='mel')
            plt.axis('off')  # Ocultar ejes
            plt.savefig(save_file, bbox_inches='tight', pad_inches=0)
            plt.close()
            
            print(f"Espectrograma generado para {filename} y guardado en la carpeta {tipo_texto}")
```
