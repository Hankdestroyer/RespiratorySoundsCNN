```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Cargar el archivo de audio
audio_path = r'C:\Users\SPARTAN PC\Documents\Trabajos\M\P\Respiratory_Sound_Database\Respiratory_Sound_Database\Frequency\106_2b1_Pl_mc_LittC2SE.wav'
y, sr = librosa.load(audio_path)

# Calcular la STFT con ventana Hanning
stft = librosa.stft(y, n_fft=2048, hop_length=512, window='hann')

# Calcular el espectrograma de potencia en dB
spectrogram_db = librosa.amplitude_to_db(np.abs(stft), ref=np.max)

# Normalizar el espectrograma a una escala lineal
normalized_spectrogram = (spectrogram_db - np.min(spectrogram_db)) / (np.max(spectrogram_db) - np.min(spectrogram_db))

# Visualizar el espectrograma original y el normalizado
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
librosa.display.specshow(spectrogram_db, sr=sr, cmap='viridis', x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Espectrograma Original')
plt.subplot(1, 2, 2)
librosa.display.specshow(normalized_spectrogram, sr=sr, cmap='viridis', x_axis='time', y_axis='log')
plt.colorbar()
plt.title('Espectrograma Normalizado')
plt.tight_layout()
plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Cargar los archivos de audio
audio_path_original = r'C:\Users\SPARTAN PC\Documents\Trabajos\M\P\Respiratory_Sound_Database\Respiratory_Sound_Database\audio_and_txt_files\106_2b1_Pl_mc_LittC2SE.wav'
audio_path_reduced = r'C:\Users\SPARTAN PC\Documents\Trabajos\M\P\Respiratory_Sound_Database\Respiratory_Sound_Database\Frequency\106_2b1_Pl_mc_LittC2SE.wav'

# Cargar los audios
y_original, sr_original = librosa.load(audio_path_original)
y_reduced, sr_reduced = librosa.load(audio_path_reduced)

# Calcular la STFT con ventana Hanning para ambos audios
stft_original = librosa.stft(y_original, n_fft=2048, hop_length=512, window='hann')
stft_reduced = librosa.stft(y_reduced, n_fft=2048, hop_length=512, window='hann')

# Calcular el espectrograma de potencia en dB
spectrogram_db_original = librosa.amplitude_to_db(np.abs(stft_original), ref=np.max)
spectrogram_db_reduced = librosa.amplitude_to_db(np.abs(stft_reduced), ref=np.max)

# Crear figura con dos subplots para comparar los espectrogramas
plt.figure(figsize=(12, 6))

# Espectrograma original
plt.subplot(1, 2, 1)
librosa.display.specshow(spectrogram_db_original, sr=sr_original, cmap='viridis', x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Espectrograma Original')

# Espectrograma reducido a 4 kHz
plt.subplot(1, 2, 2)
librosa.display.specshow(spectrogram_db_reduced, sr=sr_reduced, cmap='viridis', x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Espectrograma Reducido a 4 kHz')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.signal import spectrogram

# Ruta al archivo de audio
audio_file = r'C:\Users\SPARTAN PC\Documents\Trabajos\M\P\Respiratory_Sound_Database\Respiratory_Sound_Database\Frequency\106_2b1_Pl_mc_LittC2SE.wav'

# Cargar el archivo de audio
sample_rate, audio_data = wavfile.read(audio_file)

# Calcular el espectrograma
frequencies, times, spectrogram_data = spectrogram(audio_data, fs=sample_rate, window='hann', nperseg=256, noverlap=128)

# Redimensionar el espectrograma a 75x50 píxeles
spectrogram_resized = np.log1p(spectrogram_data)[:75, :50]  # Aplicamos logaritmo para resaltar detalles

# Mostrar el espectrograma
plt.imshow(spectrogram_resized, cmap='viridis', origin='lower', aspect='auto')
plt.colorbar(label='Log Spectrogram Amplitude')
plt.title('Resized Spectrogram')
plt.xlabel('Time Frame')
plt.ylabel('Frequency Bin')
plt.show()
```
